#### Double entry scatter plot

```julia
points = 1000
ec = EChart("scatter", rand(points), rand(points), title=Dict("text"=>"Double Scatter"))
series!(ec,"scatter",  rand(points), rand(points), name="Blue Series")
ec
```

#### Multi-series bar chart with line graph

```julia
categories = sort!([randstring(['A':'Z';], 1) for _ in range(1, stop=5)])

ec = EChart("bar",  categories, [100,130,80,50,60], resize=true)
series!(ec, "bar",  categories, [120,40,70,50,80])
series!(ec, "line", categories, [50,110,60,20,90])

ec["title"] = Dict("text"=>"Multi Series")
ec
```

#### BoxPlot and Namtso.Tools

```julia
#input data for boxplot transformation 
src = [
            [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880, 1000, 980, 930, 650, 760, 810, 1000, 1000, 960, 960],
            [960, 940, 960, 940, 880, 800, 850, 880, 900, 840, 830, 790, 810, 880, 880, 830, 800, 790, 760, 800],
            [880, 880, 880, 860, 720, 720, 620, 860, 970, 950, 880, 910, 850, 870, 840, 840, 850, 840, 840, 840],
            [890, 810, 810, 820, 800, 770, 760, 740, 750, 760, 910, 920, 890, 860, 880, 720, 840, 850, 850, 780],
            [890, 840, 780, 810, 760, 810, 790, 810, 820, 850, 870, 870, 810, 740, 810, 940, 950, 800, 810, 870],
];
```

```julia
# prepare data
prepared_src_horizontal = Namtso.prepareboxplot(src, vertical=false);
```

```julia
boxplot = EChart()
series!(boxplot, "boxplot", name = "boxplot", data = prepared_src_horizontal.boxdata)
series!(boxplot, "scatter", name = "outlier", data = prepared_src_horizontal.outliers)
boxplot.options["yAxis"] = Dict("type"=>"category", "data"=>[randstring(['A':'Z';], 1) for _ in prepared_src_horizontal.axisdata])
boxplot.options["xAxis"] = Dict("type"=>"value")

boxplot["tooltip"] = Dict("trigger"=>"item")
boxplot["width"]   = "80%"
boxplot
```

```julia
prepared_src_vertical = Namtso.prepareboxplot(src, vertical=true);

boxplot = EChart()
series!(boxplot, "boxplot", name = "boxplot", data = prepared_src_vertical.boxdata)
series!(boxplot, "scatter", name = "outlier", data = prepared_src_vertical.outliers)
boxplot.options["xAxis"] = Dict("type"=>"category", "data"=>[randstring(['A':'Z';], 1) for _ in prepared_src_vertical.axisdata])
boxplot.options["yAxis"] = Dict("type"=>"value")

boxplot["tooltip"] = Dict("trigger"=>"item")
boxplot["width"]   = "auto"
boxplot
```

#### SVG Renderer

```julia
categories = sort!([randstring(['A':'Z';], 1) for _ in range(1, stop=5)])

ec = EChart("bar",   categories, [100,130,80,50,60], resize=true, renderer=Namtso.SVG, width="110%")
series!(ec, "bar",   categories, [120,40,70,50,80])
series!(ec, "line",  categories, [50,110,60,20,90])

ec["title"] = Dict("text"=>"SVG")
ec
```
